```python
import numpy as np
# 0과 5 사이에 21개의 점을 생성  (0과 5 포함) => 총 20개의 구간 (1개 구간 = 5/20 = 0.25)
x = np.linspace(0,5,21)
# 함수 y = f(x) = x^2
y = x ** 2
# 그래프로 시각화b
plt.plot(x,y)
```
