## Iris Clistering 예제 (2018.05.24)

```python
# pandas 패키지 불러오기
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# CSV 파일을 읽어 Data Frame 변수에 저장하기
flowerData = pd.read_csv("./dataset/IRIS.csv")

# 꽃잎 길이, 넓이 기준
species = flowerData[ ['SPECIES'] ]
feature = flowerData[ ['PETALLENGTH', 'PETALWIDTH'] ]

# create model and prediction
model = KMeans(n_clusters=3, algorithm = 'auto')
model.fit(feature)
predict = pd.DataFrame(model.predict(feature))
predict.columns = ['PREDICT']

# concentrate lables to df as a new column
r = pd.concat([species, feature, predict], axis = 1)

# scatter plot (center_x, center_y, s= 50, marker = 'D', c = 'r')
plt.scatter(r['PETALLENGTH'], r['PETALWIDTH'], s=50, c = r['PREDICT'], alpha = 0.5)

plt.scatter(r['PETALWIDTH'], r['PETALLENGTH'], s=50, c = r['PREDICT'], alpha = 0.5)


#plt.scatter
print('==========')
#plt.show()
```

```python
# pandas 패키지 불러오기
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# CSV 파일을 읽어 Data Frame 변수에 저장하기
flowerData = pd.read_csv("./dataset/IRIS.csv")

# 꽃잎 & 꽃받침 길이, 넓이 기준
species = flowerData[ ['SPECIES'] ]
feature = flowerData[ ['PETALLENGTH', 'PETALWIDTH','SEPALLENGTH', 'SEPALWIDTH'] ]

# create model and prediction
model = KMeans(n_clusters=3, algorithm = 'auto')
model.fit(feature)
predict = pd.DataFrame(model.predict(feature))
predict.columns = ['PREDICT']

# concentrate lables to df as a new column
r = pd.concat([species, feature, predict], axis = 1)

# scatter plot (center_x, center_y, s= 50, marker = 'D', c = 'r')
plt.scatter(r['PETALLENGTH'], r['PETALWIDTH'], r['SEPALLENGTH'], r['SEPALWIDTH'], c = r['PREDICT'], alpha = 0.5)

#plt.scatter
plt.show()
```
